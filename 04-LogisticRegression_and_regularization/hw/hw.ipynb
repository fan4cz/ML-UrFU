## 4. Линейная регрессия + регуляризация (1 балл)
Рассмотрим задачу линейной регрессии с L2-регуляризацией. 

Пусть функция потерь имеет вид ($\frac{1}{2}$ взята для удобства дифференцирования):

$$
L(w) = \frac{1}{2} \|Xw - y\|_2^2 + \frac{\lambda}{2} \|w\|_2^2
$$


Найдём минимум этой функции. Вычислим градиент:

$$
\nabla_w L = X^T (Xw - y) + \lambda w
$$


Приравняем градиент к нулю:

$$
X^T (Xw - y) + \lambda w = 0
$$

Раскроем скобки:

$$
X^T X w - X^T y + \lambda w = 0
$$

Сгруппируем слагаемые с $w$:

$$
(X^T X + \lambda I) w = X^T y
$$

Отсюда получаем аналитическое решение — нормальное уравнение с регуляризацией:

$$
\boxed{w = (X^T X + \lambda E)^{-1} X^T y}
$$

где $E$ — единичная матрица соответствующего размера.
